Balance Statistics:

```python
balance = np.array(credit['Balance'])
plt.style.use('fivethirtyeight')
plt.hist(balance)
plt.title('Credit Card Balance Distibrution', fontsize=15)
plt.xlabel('Balance')
plt.ylabel('Number of people')
```

```python
rating = np.array(credit['Rating'])

plt.scatter(rating,balance,alpha=0.5)
plt.title('Rating vs Balance', fontsize=15)
plt.xlabel('Rating')
plt.ylabel('Balance')
```

```python
credit.corr()
```
